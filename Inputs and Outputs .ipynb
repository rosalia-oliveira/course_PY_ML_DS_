#### Pivot Tables

[Doc Pivot Tables](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html)

[Aula Pivot Tables](https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/17773168#content)

(Tabelas Dinamicas)

Permitem reorganizar dados e permitem refatorar células com base em colunas e um novo índice. De certa forma, a ideia principal é que um quaro de dados com valores repetidos dentro de suas colunas pode ser girado para reorganizacao e para uma clareza maior.

```python
df = pd.read_csv('/Users/rosaliolivei/Downloads/UNZIP_FOR_NOTEBOOKS_FINAL/03-Pandas/Sales_Funnel_CRM.csv')
df
```

```python
#Exemplo: Quantas licenças de cada tipo de produto o Google compra?
#Primeiro selecionaremos as colunas que possuem os dados dos quais iremos raspar.

licenses = df[['Company', 'Product', 'Licenses']]
licenses
```

```python
#data sao os dados filtrados, index é a coluna que esta fornecendo os valores repetidos
#neste caso é Company pois onde esta a Google que é a referencia de busca
#colunas no nosso caso sao as de licença Products

pd.pivot(data=licenses,index='Company',columns='Product', values='Licenses')
```

```python
pd.pivot_table(df, index='Company', aggfunc='sum', values=['Licenses', 'Sale Price'])
#equivale a df.groupby('Company').sum()
```

```python
pd.pivot_table(df,index=['Account Manager', 'Contact'], values=['Sale Price'], columns=['Product'
],aggfunc='sum')
```
