# DROP

<a href = 'https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html'>DOC Dataframe Drop</a>

Para NÃO APRESENTAR um coluna ou linha específica, no DataFrame podemos usar o método DROP, ele excluirá do RETORNO a coluna ou linha informada. 
Precisamos obrigatoriamente informar a coluna e o eixo(axis) - 0 = linhas e 1 = colunas (esses valores vem da função shape cfe visto em NUMPY, caso não informe o eixo, por padrão ele buscará como linhas.

```python
df.drop('tip_percent', axis = 1)
```

```python
# aqui mostra que a coluna não foi excluída permanentemente.

df.head()
```

- Caso queira <b>EXCLUIR</b> permanentemente será necessário informar o argumento inplace = True(Não Recomendado)
RECOMENDA-SE usar df = df.drop('nome_da_coluna', axis = 1) assim a coluna sera excluída permanentemente do seu arquivo de dados

### SET_INDEX


<a href = 'https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.set_index.html'>DOC Dataframe Set index</a>


Ele organiza o DataFrame baseado no seu indice rotulado, criando algo parecido com uma coluna (porém não é). Não altera o documento original, somente apresenta os dados tendo como referencia o rotulo indicado.

Para isso:

- Defina o índice DataFrame usando colunas existentes.
- Defina o índice DataFrame (rótulos de linha) usando uma ou mais colunas ou matrizes existentes. O índice pode substituir o índice existente ou expandi-lo.

```python
#df = pd.read_csv('/Users/rosaliolivei/Downloads/tips.csv')
```

```python
df.set_index('Payment ID')

#ao executarmos, não encontraremos mais a Payment ID como coluna pois agora será um índice.
```

```python
df.head()
# apenas para mostrar que o doc original não foi alterado
```

```python
# para alterar o doc original, devemos seguir a mesma recomendação do drop, que seria atribuir o comando set_index
# como valor da variavel que contem o dataframe original. Ex: df = df.set_index('Payment ID'), assim 
# haverá a alteração permanente.

df = df.set_index('Payment ID')
# ao executarmos, não encontraremos mais a Payment ID como coluna pois agora será um índice.
```

```python
df.head()
#apartir de agora, o doc original está com Payment Id como indice rotulado, permanentemente.
```

```python
#Mas ainda não o encontraremos como coluna. Até aqui, o Payment ID ainda é somente índice rotulado.

df.columns
```

#### Mas agora como obter as informações baseando-se no índice?

Podemos fazer por meio dos métodos 

<b>LOC</b>:

<a href = 'https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html'></a>

Acessa um grupo de linhas e colunas por rótulo(s) ou uma matriz booleana. É baseado principalmente em <b>rótulos</b>, mas também pode ser usado com uma matriz booleana.

OBS: o acesso se dá somente enquanto for um índice rotulado, após a transformação da informação em coluna por meio do método .reset_index(), o índice será numérico e não será possível buscar o índice desta forma. Será por através do método ILOC.

<b>ILOC</b>:

<a href = 'https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html'>Dataframe Iloc</a>

Indexação puramente baseada em localização inteira para seleção por posição. É baseado principalmente na <b>posição inteira</b> (de 0 a length-1 do eixo), mas também pode ser usado com uma matriz booleana.

```python
df.loc['Sun2959']
```

```python
#para buscar mais de uma linha usando loc

df.loc[['Sun2959', 'Sun4608']]
```

```python
#funciona para buscar pelo indice numerico também.

df.iloc[0]
```

```python
df.head()
```

```python
#para deletar uma linha, usamos o método drop porém agora usaremos o eixo 0 - axis = 0.
#Método drop para linhas se assemelha a aplicação para colunas, se for preciso, volte nas anotações e veja a diferença
#entre dropar de forma provisoria e de forma permanente.

df.drop('Sun2959', axis = 0)
```

```python
#porém se quiséssemos transformá-lo em uma coluna, podemos executar o comando abaixo:

df.reset_index()
```

```python
#Porém estamos somente alterando temporariamente e ainda não consta como uma coluna efetiva no documento.

df.columns
```

```python
#Para alterar o doc original de forma permanente, devemos atribuir o comand como valor da variável.

df = df.reset_index()

# assim consta como uma coluna de forma permanente do doc.

df.columns
```
