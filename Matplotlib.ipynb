#### Matplotlib Basics
 
[Doc Matplotlib](https://matplotlib.org/stable/plot_types/basic/index.html)

[Aula Matplotlib Basic ](https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/22972628#contentew)

O Basics refere-se aos modelos mais básicos de gráficos: lineares, dispersão(scatter), barra(bar, hastes(stem), passo(step), entre outros. Cada um deles tem uma assinatura e vários métodos. Aqui vamos ver o modelos mais básico - gráfico linear ou somente gráfico(plot).

[Doc Matplotlib plot](https://matplotlib.org/stable/plot_types/basic/plot.html#sphx-glr-plot-types-basic-plot-py)

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

```python
#primeiro tipo básico - gráfico. Onde informamos os eixos onde
#os mesmos serão plotados um versus o outro.

plt.plot(x,y)

#plt.show() opcionalmente podemos colocar ; no final do comando acima.
#obter somente a visualizaçao, sem a saída extra:
#[<matplotlib.lines.Line2D at 0x12f56e650>]
#Saida extra é um array indicando o objeto real que está sendo
#retornado, endereço de memória e a renderizacao da propria visualizacao. 
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('String title');

#podemos inserir limites na renderizacao em cada eixo:
#- .xlim, .ylim

#para salvar o grafico gerado:
#.savefig(nome.png ou nome.jpeg) se for em outro local, adicione o caminho
#antes do nome
```

#### Matplotlib Figure Objects
 
[Doc Matplotlib Figure](https://matplotlib.org/stable/api/figure_api.html)

[Aula Matplotlib Figure](https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/22972648#content)

O Figure contém todos os elementos de um gráfico. Basicamente seria nossa tela em branco, onde adicionamos os eixos e em seguida plotamos. Isso permite controles mais robustos sobre todo o contexto em que estamos desenvolvendo.

[Aula Matplotlib Figure Parameters](https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/22972672#content)

**Parameters:**
 - dpi: usado em figure() - pontos por polegada(melhora a resolucao), porém quanto mais alto, mais demorará para renderizar algo com resolucao tao alta.
 - figsize: usado em figure - tamanho da figura, recebe uma tupla onde informamos quanto por quanto queremos a figura.
 - savefig(bbox_inches)...

```python
fig = plt.figure() #criamos a tela, mas podemos personalizar as dimensoes, add os eixos
axes = fig.add_axes([0,0,1,1]) #2 primeiros parametros refere-se ao canto inferior esquerdo do eixo
#logo dos argumentos restantes referem-se ao tamanho e altura dos eixos

axes.plot(x,y);

#tente alterar os dados dos axes e veja o que acontece.

#podemos limit, title, suptitle, labels, etc em cada eixo, assim como o basics, a diferenca é:
#- .set_xtitle...
```

##### Matplotlib SubPlots

[Doc Matplotlib Subplots](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)

[Aula Matplotlib Subplots](https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/22972676#content)

Permite criar facilmente um figure object e axes em formaçoes lado a lado, retorna uma tupla e esta criara a tela da figura. 

[Doc Matplotlib subplots_adjust](https://matplotlib.org/3.2.2/api/_as_gen/matplotlib.pyplot.subplots_adjust.html)

```python
a = np.linspace(0,10,11)
b = a **4
x = np.arange(0, 10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

#axes agora é um array com 2 posicoes, para definir indidividualmente,
#valores de plot podemos faze-lo da seguinte forma:
axes[0].plot(x,y)
axes[1].plot(a,b);
```

#### Matplotlib Styling

[Aula Matplotlib Styling 01 - Legends](https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/22972864#content)

[Aula Matplotlib Styling 02 - Colors and Styles](https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/22972886#content)

Matplotlib oferece funções de estilo muito robustas que nos permitem editar todos os tipos de propriedades em nosso gráfico ou eixos: cores, legendas, adicionar marcadores, espessuras, etc.


```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend();
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) #half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00");        # RGB hex code 
```

```python
fig, ax = plt.subplots(figsize=(12,6))

# Use linewidth or lw
ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x, x-2, color="red", lw=0.50)
ax.plot(x, x-3, color="red", lw=1)
ax.plot(x, x-4, color="red", lw=10);
```

```python
# possible linestype options ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--'); # dashes
```

```python
# https://matplotlib.org/3.2.2/api/markers_api.html

fig, ax = plt.subplots(figsize=(12,6))

# Use marker for string code
# Use markersize or ms for size

ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) #ms can be used for markersize
ax.plot(x, x-3,marker='s',ms=20,lw=0) # make linewidth zero to see only markers
ax.plot(x, x-4,marker='1',ms=20);
```

#### Tarefa Um: Criar dados a partir de uma equação

É importante ser capaz de traduzir directamente uma equação real numa parcela. A sua primeira tarefa é, na realidade, puramente entorpecida, depois exploraremos a forma de a traçar com Matplotlib. A [equação mundialmente famosa](https://en.wikipedia.org/wiki/Mass%E2%80%93energy_equivalence) de Einstein:

$$E=mc^2$$


Use os seus conhecimentos de Numpy para criar duas matrizes: E e m , onde **m*** é simplesmente 11 valores uniformemente espaçados representando 0 gramas a 10 gramas. E deve ser a energia equivalente para a massa. Terá de descobrir o que deve fornecer **c*** para as unidades m/s, uma pesquisa rápida no Google dar-lhe-á facilmente a resposta (utilizaremos a aproximação próxima nas nossas soluções).

Traduzido com a versão gratuita do tradutor - www.DeepL.com/Translator

```python
m = np.linspace(0,10,11)
m
```

```python
#velocidade da luz é aproximadamente 3 x 10ˆ8
c = 3 * (10 ** 8)
c
```

```python
E= m*c**2
E
```

```python
import matplotlib.pylab as plt
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(m, E, color="red", linewidth=4)
axes.set_xlim(0,10);
axes.set_xlabel('Energy in Joules')
axes.set_ylabel('Mass in Grams')
axes.set_title('E=mcˆ2');
```
