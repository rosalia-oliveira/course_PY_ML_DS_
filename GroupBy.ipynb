Agrupa DataFrame usando um mapeador ou por uma série de colunas.

Uma operação groupby envolve alguma combinação de divisão do objeto, aplicação de uma função e combinação dos resultados. Isso pode ser usado para agrupar grandes quantidades de dados e operações de computação nesses grupos. Nos permite examinar dados por categoria.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/rosaliolivei/Downloads/UNZIP_FOR_NOTEBOOKS_FINAL/03-Pandas/mpg.csv')
df
```

```python
#Agrupamento pela coluna model_year, porém trazendo a média de mpg
#Numeric Only foi adicionado para excluir o warning conforme link abaixo: 
#https://stackoverflow.com/questions/72223610/dropping-invalid-columns-futurewarning

#mpg - milhas por galao

df.groupby('model_year').mean(numeric_only=True)['mpg']
```

```python
#Mas ao invés de agrupar por uma unica coluna, poderiamos incluir para uma coluna no agrupamento. Sera que é possivel?

df.groupby(['model_year', 'cylinders']).mean(numeric_only=True)

#percebe-se que agora temos index múltiplo e podemos fazer algumas análises, 
#por exemplo, modelo do ano 1970 4 cilindros tem uma media maior de desempenho que o
#de 8 cilindros, etc.
```

```python
df.groupby(['model_year', 'cylinders']).mean(numeric_only=True).index
```

```python
year_cyl = df.groupby(['model_year', 'cylinders']).mean(numeric_only=True)

year_cyl.xs(key=70, level='model_year')

#para buscar todos os dados de mais de uma key, devemos usar o loc:
# year_cyl.loc[[70,80]]
```

```python
#Mas se quisermos buscar um cilindro especifico porem de todos os anos?
#Devemos usar o xs
#exemplo: quero todas as instancias de 4 cilindros para todos os anos.

year_cyl.xs(key=4, level='cylinders')
```

```python
#E se quisermos buscar as informacoes de cilindros 6 e 8 do ano 70?

df[df['cylinders'].isin([6,8])].groupby(['model_year', 'cylinders']).mean(numeric_only=True)
```

```python
#Trocar a ordem dos index, sem mudar os valores

year_cyl #index 1 - model_year, index 2 - cylinders
```

```python
year_cyl.swaplevel() #index 1 - cylinders, index 2 - model_year
```

```python
#sempre ter em mente que é recomendável ordenar sempre pelo index mais externo, 
#teste o comando abaixo em seu ambiente alterando model_year para cylinders. 


year_cyl.sort_index(level = 'model_year', ascending = True)
```

```python
#O método agg() é usado quando precisamos aplicar mais de uma operacao sobre o eixo especificado.

df.agg(['std', 'mean'])
df.agg(['std', 'mean'])['mpg']
```

```python
#Métodos agregados por coluna - passar em um dicionário onde as chaves são as colunas e os valores são uma lista.
df.agg({'mpg':['median','mean'],'weight':['mean','std']})
```

```python
#Com GroupBy

df.groupby('model_year').agg({'mpg':['median','mean'],'weight':['mean','std']})
```
