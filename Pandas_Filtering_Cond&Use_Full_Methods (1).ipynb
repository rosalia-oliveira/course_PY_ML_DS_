#### Filtering Conditional

Já vimos como filtrar colunas e linhas com base em sua posição ou índice rotulado. No entanto, normalmente na análise de dados, nossos conjuntos de dados serão grandes o suficiente para que não filtremos com base em uma posição.

#### Vamos aplicar um filtro baseado em duas ou mais condições

Para aplicar multiplos filtros podemos utilizar:

- & (E) (que corresponde ao AND em Python ou && em outras linguagens) - neste caso, as duas condições devem ser verdadeiras.
- | (OU) (que corresponde ao OR em Python ou || em outras linguagens) - neste caso, basta uma das condições serem verdadeiras.

```python
#Vamos filtrar por sexo e se é fumante sex = male e smoke = yes
    
df[(df['sex'] == 'Male') & (df['smoker'] == 'Yes')]
```

```python
#Vamos filtrar por valor total da conta ou se é masculino conta > 30 ou sex = male
    
df[(df['total_bill'] > 30) | (df['sex'] == 'Female')]
```

```python
#Vamos filtrar mais um exemplo, onde so nos importa as informacoes de ocorrencias no sabado ou domingo.

df[(df['day'] == 'Sun') | (df['day'] == 'Sat')]
```

```python
#Mas se tivéssemos que adicionar mais condições, logo ia começar a ficar mais difícil de compreender,
#para resolver isso, podemos usar o método isin(). --> esta dentro. Vamos ao exemplo:

df[df['day'].isin(['Sat', 'Sun'])]
```

O primeiro e talvez o mais utilizados de todos, o que ele faz é permitir que você aplique qualquer função personalizada do seu próprio Python a cada linha em uma série, para que possamos usar uma ou várias colunas como entrada.

Neste momento, usaremos exemplo baseando-se em somente UMA coluna. E a medida que avançamos, vamos incluindo para multiplas colunas.

```python
#Vamos ao nosso dataframe 

df.head()
```

```python
#Temos p CCNumber e queremos saber buscar somente os 4 ultimos digitos do mesmo.

cc_number = df['CC Number'][0]
cc_number
```

```python
#vamos tentar pegar o primeiro digito sem transformar em string - descomente este exemplo e teste cc_number[0]
# ele acusará um erro
# IndexError: invalid index to scalar variable

#cc_number[0] 

str(cc_number)[0] # resultado 3

#mas se quisessemos buscar os 4 ultimos digitos:

str(cc_number)[-4:] # numeros negativos indicam para iniciar a busca pelo fim.
```

```python
#Porem o Pandas não possui um método embutido para executar o processo acima, o str que usamos vem do Python.
#Para isso usaremos o método apply passando para ele uma função que retornará os dados que queremos.
# primeiro criaremos a função:

def last_four(cc):
    return str(cc)[-4:]

#agora localizamos a coluna que sera trabalhada e submetemos cada valor a funcao com o método apply:

df['CC Number'].apply(last_four) 
#note que nao estamos chamando ou executando a função, estamos apenas informando
#a funcao que sera utilizada e o pandas se encarregara se executa-la com o valor de cada linha.
```

```python
#poderiamso facilmente agora criar uma nova coluna e atribuir esta serie como valor, exemplo:
#df['last_four'] = df['CC Number'].apply(last_four)
```

```python
#Vamos a outro exemplo. Temos a coluna de total da conta e queremos que retorne o numero de 
#cifroes referentes as dezenas, por exemplo, 10.00 retorna $, 20.00 $$ e assim por diante.

#vamos construir a função e submete-la ao metodo apply.

def yelp(price):
    if price < 10:
        return '$'
    elif price < 20 and price >=10:
        return '$$'
    else: 
        return '$$$'
```

```python
df['total_bill'].apply(yelp)
```

##### Método APPLY on Multiple Column

<a href = 'https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/22971386#content'>Aula Metodo Apply On Multiple Column</a>

<a href = 'https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html'>Método Apply</a>

Muitas vezes precisaremos fazer o cruzamento dos dados de duas ou mais colunas. as chamdas múltiplas entradas, e podemos utilizar o apply também nestes casos.

Primeiro vamos aprender a converter uma função em um expressão <b>Lambda</b>, mas o que significa isso?

São chamadas de funções anônimas, são do que funções que o usuário não precisa definir, ou seja, não vai precisar escrever a função e depois utilizá-la dentro do código.

```python
#Vamos a um exemplo:

def simple_fn(num):
    return num*2

#Vamos reescreve-la como lambda

lambda num: num*2
```

```python
#Ta agora vamos trabalhar com a ideia de ter um percentual de gorjeta.
#Vamos criar uma coluna tip_percent(como ja fizemos em outro exemplo), porém para calcular este percentual nós
#precisaremos prever multiplas entradas total_bill e tip.
#De acordo com o percentual, iremos rotular se a gorjeta é generosa.

#Vamos primeiro criar uma função:

def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
#Agora vamos usar uma lambda como funcao para o apply.
#primeiro pegaremos as colunas que precisaremos e depois aplicaremos o apply.

df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis = 1)
```

```python
df.head()
```

```python
#Para otimizar vamos usar um método do Numpy que além de melhorar o tempo computacional 
#ainda tem uma sintaxe mais simples. É o VECTORIZE.

df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])

#nome da nova coluna = método(função a ser chamada e executada)(colunas)
```

```python
df.head()
```

##### Método NLargest

<a href = 'https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.nlargest.html'>Doc Nlargest</a>

Retorna as primeiras n linhas com os maiores valores nas colunas, em ordem decrescente. As colunas não especificadas também são retornadas, mas não são usadas para ordenação.

Este método é equivalente a df.sort_values(columns, ascending=False).head(n), mas com melhor desempenho.

Retorna as primeiras n linhas com os menores valores nas colunas, em ordem crescente. As colunas não especificadas também são retornadas, mas não são usadas para ordenação.

Este método é equivalente a df.sort_values(columns, ascending=True).head(n), mas com melhor desempenho.
