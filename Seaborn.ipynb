##### Distribution Plots

[Aula01 Distribution Plots](https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/22975834#content)

[Aula02 Distribution Plots](https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/22975840#overview)

Gráficos de distribuiçao exibem um único recurso cont;inuo e ajudam a visualizar propriedades, como desvio e valores médios.

Existem tres tipos principais:

- **Rug Plot**: tipo mais simples de grafico de distribuicao, o que ele faz é adicionar um unico traço ou linha de marcação para cada valor. Uma particularidade é que o eixo Y não é realmente interpretável, ele não tem muito sentido

[DOC Rug Plot](https://seaborn.pydata.org/generated/seaborn.rugplot.html)


- **Histograma**: representa a visualizacao de clássica de distribuiçao de uma ou mais variaveis contando o numero de observacoes que se enquandram em compartimentos discretos. É uma espécie de gráfico de barras que demonstra uma distribuiçao de frenquencias. Sua base representa uma classe e a altura representa a quantidade ou frequencia absoluta com uqe o valor de cada classe ocorre.

[DOC histrogram](https://seaborn.pydata.org/generated/seaborn.histplot.html)

[DOC Displot](https://seaborn.pydata.org/generated/seaborn.displot.html)
OBS: disTplot está depreciado.

- KDE Plot - Kernel Density Estimation: método para visualizar a distribuiçao de observaçoes em conjunto de dados, análogo a um histograma. Representa os dados usando uma curva de densidade de probabilidade contínua em uma ou mais dimensoes e também pode traçar um unico gráfico para várias amostras o que ajuda a produzir um gráfico menos confuso e mais interpretável.

[DOC KDE Plot](https://seaborn.pydata.org/generated/seaborn.kdeplot.html)

```python
sns.rugplot(x='salary', data=df);
```

```python
sns.displot(data=df, x='salary', bins=20, color='pink', edgecolor='yellow', kde=True, rug=True);

#bins define a quantidade de colunas que deve ser apresentada
#explorsns.set(ver assinatura) para alterar o estilo de plano de fundo.
```

```python
sns.histplot(data=df, x='salary', bins=20, color='pink', edgecolor='yellow');
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sns.rugplot(x='age', data=sample_ages);
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=20, kde=True);
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100])
```

## Categorical Plots

[Aula introdutoria](https://meli.udemy.com/course/python-for-machine-learning-data-science-masterclass/learn/lecture/22975850#content)

Gráficos categóricos sao usados praticamente para visualizar a relacao entre as variáveis. Elas podem ser totalmente numéricas ou uma categoria como um grupo, classe ou divisão. Existem dois tipos principais que sao:

- **Count Plot**: gráficos que simplesmente contam o número de linhas por categoria, a mesma coisa que fazer um GROUP BY e em seguida chamar a contagem;

[DOC Count Plot](https://seaborn.pydata.org/generated/seaborn.countplot.html)

- **Bar Plot**: representa uma estimativa de tendencia central para uma variavel numerica com a acltura de cada retangulo e fornece alguma indicaçao da incerteza em torno dessa estimativa usando barras de erro.

[DOC Bar Plot](https://seaborn.pydata.org/generated/seaborn.barplot.html)
